```python
import numpy as np
import seaborn as sns

#create the array based on number values
def update_array(arr, value):
    size = arr.shape[0]
    #each block gets smaller by a factor of four
    quart_size = size // (4 ** value)
    
    #iterate over the grid with a repeating pattern of 4
    for a in range(4):
        #locates the upper right corner of each square and colors it
        for i in range(quart_size * a, size, 4 * quart_size):
            for j in range(quart_size * a, size, 4 * quart_size):
                #creates squares of appropriate size based on the upper-right corner location
                arr[i:i+quart_size, j:j+quart_size] += 1
    return arr


def run(limit):
    size = 4 ** limit
    array = np.zeros((size, size))
    for i in range(limit):
        array = update_array(array, i + 1)
    # create a heatmap using Seaborn
    colors = sns.cubehelix_palette(start=2.1, rot=.1, dark=.95, light=.85-(.1*limit), as_cmap=True)
    sns.heatmap(array, annot=False, cmap=colors, cbar=False, square=True, xticklabels=False, yticklabels=False)

#call the function, this is pattern 1 in the puzzle example
run(3)
```

```python

run(4)
```
